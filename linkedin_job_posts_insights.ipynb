```python
import pandas as pd

# Load the Excel file
df = pd.read_excel("linkedin_job_posts_insights.xlsx")

# Display basic info about the dataset
print(" Dataset loaded successfully!")
df.head()
```

```python
#Basic data explorations
# Check shape, columns, and datatypes
df.info()

# Basic stats (works only for numeric columns)
df.describe(include='all')

# Check for null values
df.isnull().sum()
```

```python
#cleaning the dataset to derive meaning ful insights
# Drop duplicates
df = df.drop_duplicates()

# Handle missing values
df = df.dropna(subset=['job_title', 'company_name', 'location'])

print("Cleaned dataset shape:", df.shape)
```

```python
#analyze most common job titles
top_titles = df['job_title'].value_counts().head(10)
print(" Top 10 Job Titles:")
print(top_titles)
```

```python
#anayze top hiring locations
top_locations = df['location'].value_counts().head(10)
print("\n Top 10 Locations:")
print(top_locations)
```

```python
#analyze industries
top_industries = df['industry'].value_counts().head(10)
print("\n Top 10 Industries:")
print(top_industries)
```

```python
#Analyze seniority levels
top_seniority = df['seniority_level'].value_counts()
print("\n Seniority Level Distribution:")
print(top_seniority)
```

```python
#Analyze employment type
top_employment = df['employment_type'].value_counts()
print("\n Employment Type Distribution:")
print(top_employment)
```

```python
#visualizations
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
top_titles.plot(kind='barh', title='Top 10 Job Titles')
plt.xlabel("Number of Postings")
plt.show()

plt.figure(figsize=(8,4))
top_locations.plot(kind='barh', title='Top 10 Locations')
plt.xlabel("Number of Postings")
plt.show()

plt.figure(figsize=(6,4))
top_seniority.plot(kind='bar', title='Seniority Levels')
plt.ylabel("Count")
plt.show()
```

```python
#viewing insights
print(" Summary of Insights:")
print("Most common job titles:\n", top_titles.index.tolist())
print("Top hiring locations:\n", top_locations.index.tolist())
print("Top industries:\n", top_industries.index.tolist())
print("Seniority levels:\n", top_seniority.to_dict())
print("Employment types:\n", top_employment.to_dict())
```
